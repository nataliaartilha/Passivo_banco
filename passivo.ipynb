```python
def calcular_dias_uteis(data_inicio, data_fim):
    dias_uteis = 0
    data_atual = data_inicio

    while data_atual < data_fim:
        if data_atual.weekday() < 5 and data_atual not in feriados10:
            dias_uteis += 1
        data_atual += timedelta(days=1)

    return dias_uteis

base_passivo['Dias_Uteis2'] = base_passivo.apply(lambda row: calcular_dias_uteis(row['Data_ref'], row['Vencto']), axis=1)

base_passivo.head()
```
